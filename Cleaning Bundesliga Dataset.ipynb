```python
def map_free_move_loan(x):
    if type(x) == str:
        if "ablösefrei" in x:
            return 2
        elif "Leih" in x:
            return 1
        else: pass
    else:
        pass
```

```python
df.loc[df.transfer_type.isnull(), "transfer_type"] = df.loc[df.transfer_type.isnull()].transfer_fee.map(map_free_move_loan)
```

```python
def map_fee(x):
    if type(x) == str:
        if "Leih" in x:
            return 0
        elif "Tsd" in x:
            return float(x[:-7].replace(",", ".")) / 1000
        elif "Mio" in x:
            return float(x[:-7].replace(",", "."))
        else:
            return x
    else:
        return x
```

```python
df.loc[df.transfer_type.isnull(), "transfer_type"] = 0
```

```python
df["transfer_fee"] = df.transfer_fee.map(map_fee)
```
